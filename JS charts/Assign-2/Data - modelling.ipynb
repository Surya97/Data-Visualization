```python
player_stats_dict = {}
```

```python
final_winners = []
```

```python
for index, row in df.iterrows():
#     print(index)
    if row['round'] == "Final":
        final_winners.append(row['winner'])
    for i in range(1, 3):
        player_ = "player"+str(i)
        firstServe_ = "firstServe"+str(i)
        firstPointWon_ = "firstPointWon"+str(i)
        secondPointWon_ = "secPointWon"+str(i)
        break_ = "break"+str(i)
        return_ = "return"+str(i)
        ace = "ace"+str(i)
        double = "double"+str(i)
        error = "error"+str(i)
#         net_ = "net"+str(i)
        
        firstServe = row[firstServe_].replace('%','').replace('-', '')

        firstPointWon = row[firstPointWon_].replace('%','').replace('-', '')

        secPointWon = row[secondPointWon_].replace('%','').replace('-','')

        breakPoint = row[break_].replace('%','').replace('-', '')

        returnPoint = row[return_].replace('%','').replace('-', '')
        
        ace_ = row[ace]
        
        double_ = row[double]
        
        error_ = row[error]
        
        if np.isnan(ace_):
            ace_ = 0
        if np.isnan(double_):
            double_ = 0
        if np.isnan(error_):
            error_ = 0
        
#         net = row[net_].replace('%','').replace('-', '')
        
        
#         print(firstServe, firstPointWon, secPointWon, breakPoint, returnPoint, net)
        
        if firstServe == '':
            firstServe = '0'
        if firstPointWon == '':
            firstPointWon = '0'
        if breakPoint == '':
            breakPoint = '0'
        if returnPoint == '':
            returnPoint = '0'
        if secPointWon == '':
            secPointWon = '0'
#         if net == '':
#             net = '0'
        player = row[player_]
        
        if row[player_] in player_stats_dict:
#             player_stats_dict[player]['firstServe'].append(int(firstServe))
            player_stats_dict[player]['firstPointWon'].append(int(firstPointWon))
            player_stats_dict[player]['secPointWon'].append(int(secPointWon))
            player_stats_dict[player]['break'].append(int(breakPoint))
            player_stats_dict[player]['return'].append(int(returnPoint))
            player_stats_dict[player]['ace'].append(ace_)
            player_stats_dict[player]['double'].append(double_)
            player_stats_dict[player]['error'].append(error_)
#             player_stats_dict['net'].append(int(net))
            player_stats_dict[player]['matches'] = player_stats_dict[player]['matches'] + 1
            
            
            
        else:
            player_stats_dict[player] = {'champion':0}
#             player_stats_dict[player]['firstServe'] = [int(firstServe)]
            player_stats_dict[player]['firstPointWon'] = [int(firstPointWon)]
            player_stats_dict[player]['secPointWon'] = [int(secPointWon)]
            player_stats_dict[player]['break'] = [int(breakPoint)]
            player_stats_dict[player]['return'] = [int(returnPoint)]
            player_stats_dict[player]['ace'] = [ace_]
            player_stats_dict[player]['double'] = [double_]
            player_stats_dict[player]['error'] = [error_]
#             player_stats_dict['net'] = [int(net)]
            player_stats_dict[player]['matches'] = 1
            
        if row['round'] == "Final" and row['winner'] == player:
                player_stats_dict[player]['champion']+=1
    
        
```

```python
for key, value in player_stats_dict.items():
    print(key, value)
    break
```

```python
player_stats_dict_aggregate = {}
```

```python
for key, value in player_stats_dict.items():
    
    if key not in player_stats_dict_aggregate:
        player_stats_dict_aggregate[key] = {}
        
#     player_stats_dict_aggregate[key]['firstServe'] = sum(value['firstServe'])/value['matches']
    player_stats_dict_aggregate[key]['firstPointWon'] = sum(value['firstPointWon'])/value['matches']
    player_stats_dict_aggregate[key]['secPointWon'] = sum(value['secPointWon'])/value['matches']
    player_stats_dict_aggregate[key]['break'] = sum(value['break'])/value['matches']
    player_stats_dict_aggregate[key]['return'] = sum(value['return'])/value['matches']
    player_stats_dict_aggregate[key]['ace'] = max(value['ace'])
    player_stats_dict_aggregate[key]['double'] = max(value['double'])
    player_stats_dict_aggregate[key]['error'] = sum(value['error'])/value['matches']
    player_stats_dict_aggregate[key]['champion'] = value['champion']
    
```

```python
for key, value in player_stats_dict_aggregate.items():
    print(key, value)
    break
```
